```python
from tkinter import *
root = Tk()
root.title("Database")

# User entry
previous_reading = Entry(root, width=40, borderwidth=5)
current_reading = Entry(root, width=40, borderwidth=5)
gcv = Entry(root, width=40, borderwidth=5)
factor = Entry(root, width=40, borderwidth=5)


# get values
def reading():
    global previous_reading, current_reading, gcv, factor
    
    previous_reading = int(previous_reading.get())
#    previous_reading = int(previous_reading)
    current_reading  = current_reading.get()
    current_reading = int(current_reading)
    gcv = gcv.get()
    gcv = int(gcv)
    factor = factor.get()
    factor = float(factor)
    print(f"Current reading: {current_reading}, \nPrevious reading: {previous_reading}, \nGCV: {gcv}, \nPress/Factor: {factor}", sep="\n") 

    
# Calculation 
    difference = current_reading - previous_reading
    #factor = float(input("Factor: "))
    #gcv = 1057
    fix = 281.7385
    hm3 = difference * factor / 100000
    mmbtu = hm3 * gcv / fix
    meter_rent = 20.67
    if mmbtu <= 0.5:
        price = 121
        total = hm3*price
        gst = total * 17/100
        print("Amount:",total+gst+meter_rent)
    elif mmbtu <= 1.00:
        price = 300
        total = hm3*price
        gst = total * 17/100
        print("Amount:",total+gst+meter_rent)
    elif mmbtu <= 2.00:
        price = 553
        total = hm3*price
        gst = total * 17/100
        print("Amount:",total+gst+meter_rent)
    elif mmbtu <= 3.00:
        price = 738
        total = hm3*price
        gst = total * 17/100
        print("Amount:",total+gst+meter_rent)
    elif mmbtu <= 4.00:
        price = 1107
        total = hm3*price
        gst = total * 17/100
        print("Amount:",total+gst+meter_rent)
    elif mmbtu > 4.00:
        price = 1460
        total = hm3*price
        gst = total * 17/100
        print("Amount:",total+gst+meter_rent)
    else:
        print("Amount is zero")


# set position of user entry
previous_reading.grid(row=0, column=1, padx=10, pady=10)
current_reading.grid(row=1, column=1, padx=10, pady=10)
gcv.grid(row=2, column=1, padx=10, pady=10)
factor.grid(row=3, column=1, padx=10, pady=10)

# create lables

lable_previous_reading = Label(root, text = "Previous_reading: ")
lable_current_reading = Label(root, text = "Current_reading: ")
lable_gcv = Label(root, text = "GCV: ")
lable_factor = Label(root, text = "Pres//Factor: ")

lable_previous_reading.grid(row=0, column=0, padx=10, pady=10)
lable_current_reading.grid(row=1, column=0, padx=10, pady=10)
lable_gcv.grid(row=2, column=0, padx=10, pady=10)
lable_factor.grid(row=3, column=0, padx=10, pady=10)

# result box
result_box_label = Label(root, text = "Amount: ")
result_box_label.grid(row=4, column=0, padx=10, pady=10)
result_box = Text(root, width=20, height=1, borderwidth=5, fg="blue")
result_box.grid(row=4, column=1, padx=10, pady=10)

# create submit button
    
button_sub = Button(root, text="Calculate", padx=10, pady=10, command= reading)
button_sub.grid(row=5, column=0, columnspan = 2)


root.mainloop()

```
